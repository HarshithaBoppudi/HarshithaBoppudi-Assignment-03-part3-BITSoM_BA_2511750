** Exception Handling**

```python
# Guarded Calculator
def safe_divide(a, b):
    try:
        # Attempt division
        result = a / b
        return result
    except ZeroDivisionError:
        # Handle division by zero
        return "Error: Cannot divide by zero"
    except TypeError:
        # Handle invalid input types (like strings)
        return "Error: Invalid input types"

print(safe_divide(50, 2))      # Normal division
print(safe_divide(10, 0))      # Division by zero
print(safe_divide("ten", 2))   # Invalid input type
```

```python
# Guarded File Reader:
def read_file_safe(filename):
    try:
        with open(filename, "r") as f:
            content = f.read()
            return content
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
        return None
    finally:
        print("File operation attempt complete.")

# --- Testing the function ---
# Case 1: File exists
print("\n--- Reading python_notes.txt ---")
result1 = read_file_safe("python_notes.txt")
if result1:
    print("File content:\n", result1)

# Case 2: File does not exist
print("\n--- Reading ghost_file.txt ---")
result2 = read_file_safe("ghost_file.txt")
if result2:
    print("File content:\n", result2)
```

```python
# Robust API Calls:
import requests

# Utility function to safely perform GET requests
def safe_get(url, timeout=5):
    try:
        response = requests.get(url, timeout=timeout)
        return response.json()
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print("Unexpected error:", e)

# Utility function to safely perform POST requests
def safe_post(url, json_data, timeout=5):
    try:
        response = requests.post(url, json=json_data, timeout=timeout)
        return response.json()
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print("Unexpected error:", e)

# Step 1 — Fetch 20 products
data = safe_get("https://dummyjson.com/products?limit=20")
if data and "products" in data:
    products = data["products"]
    print(f"{'ID':<4} | {'Title':<30} | {'Category':<15} | {'Price':<8} | {'Rating'}")
    print("-"*75)
    for p in products:
        print(f"{p['id']:<4} | {p['title']:<30} | {p['category']:<15} | ${p['price']:<7.2f} | {p['rating']}")

    # Step 2 — Filter and Sort
    filtered = [p for p in products if p['rating'] >= 4.5]
    sorted_products = sorted(filtered, key=lambda x: x['price'], reverse=True)
    print("\nFiltered and Sorted Products (rating >= 4.5):")
    for p in sorted_products:
        print(f"{p['title']} | ${p['price']} | Rating: {p['rating']}")

# Step 3 — Search by Category (laptops)
laptops_data = safe_get("https://dummyjson.com/products/category/laptops")
if laptops_data and "products" in laptops_data:
    print("\nLaptops Found:")
    for l in laptops_data["products"]:
        print(f"{l['title']} | ${l['price']}")

# Step 4 — POST Request (Simulated)
new_product = {
    "title": "My Custom Product",
    "price": 999,
    "category": "electronics",
    "description": "A product I created via API"
}
post_response = safe_post("https://dummyjson.com/products/add", new_product)
if post_response:
    print("\nPOST Response:")
    print(post_response)
```

```python
#  Input Validation Loop:
import requests

while True:
    user_input = input("Enter a product ID to look up (1–100), or 'quit' to exit: ")

    # Step 1: Exit condition
    if user_input.lower() == "quit":
        print("Exiting program.")
        break

    # Step 2: Validate input is an integer in range 1–100
    if not user_input.isdigit():
        print("Warning: Please enter a valid integer between 1 and 100.")
        continue

    product_id = int(user_input)
    if product_id < 1 or product_id > 100:
        print("Warning: Product ID must be between 1 and 100.")
        continue

    # Step 3: Make GET request safely
    url = f"https://dummyjson.com/products/{product_id}"
    try:
        response = requests.get(url, timeout=5)

        # Step 4: Check HTTP status codes
        if response.status_code == 404:
            print("Product not found.")
        elif response.status_code == 200:
            product = response.json()
            print(f"Product found: {product['title']} | Price: ${product['price']}")
        else:
            print(f"Unexpected HTTP status: {response.status_code}")

    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print("Unexpected error:", e)
```

```python
import requests
from datetime import datetime

# Utility function to log errors
def log_error(context, error_type, message):
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    entry = f"[{timestamp}] ERROR in {context}: {error_type} — {message}\n"
    with open("error_log.txt", "a", encoding="utf-8") as log_file:
        log_file.write(entry)

# --- Trigger 1: ConnectionError ---
try:
    # Intentionally unreachable host
    response = requests.get("https://this-host-does-not-exist-xyz.com/api", timeout=5)
except requests.exceptions.ConnectionError:
    log_error("fetch_products", "ConnectionError", "No connection could be made")

# --- Trigger 2: HTTPError (404) ---
try:
    response = requests.get("https://dummyjson.com/products/999", timeout=5)
    if response.status_code == 404:
        log_error("lookup_product", "HTTPError", "404 Not Found for product ID 999")
    elif response.status_code == 200:
        product = response.json()
        print(f"Product found: {product['title']} | Price: ${product['price']}")
except requests.exceptions.ConnectionError:
    log_error("lookup_product", "ConnectionError", "No connection could be made")
except requests.exceptions.Timeout:
    log_error("lookup_product", "Timeout", "Request timed out")
except Exception as e:
    log_error("lookup_product", "UnexpectedError", str(e))

# --- Read and print the log file ---
print("\nContents of error_log.txt:")
with open("error_log.txt", "r", encoding="utf-8") as log_file:
    print(log_file.read())
```
